```python
import matplotlib.pyplot as plt

days = ["Mon", "Tue", "Wed", "Thu", "Fri", "Sat", "Sun"]

sleep = [7, 8, 5, 6, 7, 7, 8]
work = [9, 8, 7, 9, 4, 0, 0]
study = [3, 2, 3, 4, 3, 2, 4]
spiritual = [2.5, 2, 2.5, 2, 2.5, 2, 2.5]
others = [2, 2, 2, 2, 2, 6, 8]  # meals, travel, misc

plt.figure(figsize=(10,6))
plt.stackplot(
    days,
    sleep,
    work,
    study,
    leisure,
    others,
    labels=["Sleep", "Work", "Study", "Leisure", "Others"]
)

plt.legend(loc="upper left")
plt.title("My Weekly Time Distribution")
plt.ylabel('Hours Per Day')
plt.xlabel("Days")
```
