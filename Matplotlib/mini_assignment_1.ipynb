```python
import matplotlib.pyplot as plt

years = list(range(2014, 2024))

kohli_runs = [1100, 1020, 980, 1050, 1200, 1150, 900, 870, 1300, 1250]
rohit_runs = [900, 950, 1000, 980, 1100, 1080, 850, 920, 1200, 1180]
sehwag_runs = [850, 800, 780, 760, 740, 700, 650, 600, 580, 550]

plt.style.use("ggplot")   # one custom style
plt.figure(figsize=(10, 5))

plt.plot(
    years, kohli_runs,
    color="blue",
    linestyle="-",
    marker="o",
    label="Virat Kohli"
)

plt.plot(
    years, rohit_runs,
    color="green",
    linestyle="--",
    marker="s",
    label="Rohit Sharma"
)

plt.plot(
    years, sehwag_runs,
    color="red",
    linestyle=":",
    marker="^",
    label="Virender Sehwag"
)

plt.xlabel("Years")
plt.ylabel("Runs")
plt.title("Hypothetical Runs Comparison (20142023)")

plt.legend()
plt.tight_layout()
plt.savefig("runs_comparison.png")
plt.show()

```
