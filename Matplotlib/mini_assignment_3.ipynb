```python
import matplotlib.pyplot as plt
```

```python
os = ["Android", "iOS", "Others"]
market_share = [72, 26, 2]
```

```python
plt.figure()
plt.pie(
    market_share,
    labels = None,
    autopct='%1.1f%%',
    startangle=90,
    pctdistance=1.2   # ðŸ”‘ pushes % text outside
)
plt.legend(os, loc = "center left", bbox_to_anchor=(1,0.5))
plt.title("Mobile Os Market Share")
plt.tight_layout()
```

# horizonta Bar Chart

```python
plt.figure()
plt.barh(os, market_share)
plt.xlabel("Market Share (%)")
plt.title("Mobile OS Market Share")
```
