# PART 1️ Side-by-Side Bar Chart (5 Selected Years)

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
years = ["2005", "2008", "2011", "2014", "2017"]

sachin_runs = [900, 1100, 1200, 800, 0]
kohli_runs  = [0, 700, 900, 1300, 1400]
sehwag_runs = [1100, 1200, 1000, 700, 0]
```

```python
x = np.arange(len(years))
width = 0.25
```

```python
plt.figure(figsize=(10, 5))
plt.bar(x - width, sachin_runs, width=width, label="Sachin", color="blue")
plt.bar(x, kohli_runs, width=width, label="Kohli", color="orange")
plt.bar(x + width, sehwag_runs, width=width, label="Sehwag", color="purple")

plt.xticks(x, years)
plt.xlabel("Year")
plt.ylabel("Runs")
plt.title("Runs Comperision Over Selected Years")
plt.legend()
plt.tight_layout()
```

# Part 2 Horizontal Bar Chart – Total Runs in Debut 5 Years

```python
players = ["Sachin", "Kohli", "Sehwag"]
debut_5yr_runs = [2800, 4200, 3500]

plt.figure(figsize=(8, 4))

plt.barh(players, debut_5yr_runs, color=["#0759e8", "#07e8bf", "#7b07e8"])

plt.xlabel("Total Runs (First 5 Years)")
plt.title("Total Runs in Debut 5 Years")

plt.tight_layout()

```
