```python
import requests

url = "https://books.toscrape.com/"
response = requests.get(url)

for i in range(1, 51):
    a = requests.get(f"https://books.toscrape.com/catalogue/page-{i}.html")
    with open(f"Html/page{i}.html","w",encoding="utf-8")as f:
        f.write(a.text)
        print(f"Downloaded file {i} ")
```

```python
from bs4 import BeautifulSoup
with open("Html/page1.html") as f:
    content = f.read()
```

```python
soup = BeautifulSoup(response.text, "html.parser")
```

```python
h3s = soup.find_all("h3")
```

```python
articles = soup.select("article.product_pod")
```

```python
items = []
for article in articles:
    title = article.find("h3").find("a")["title"]
    price = article.select_one("p.price_color").text.split("Â£")[1]
    rating_element = article.select_one("p.star-rating")
    rating = rating_element['class'][1]
    items.append([title,price,rating])
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(items,columns=["Books Title","Price In Euro","Ratings"])
df
```

```python
df.to_csv("Data.csv",index=False)
```
