```python
import pandas as pd
import numpy as np

df = pd.read_csv("../Data/Processed/cleaned_job_listings.csv")
```

```python
df["ExperienceLevel"] = np.select(
    [
        df["Exp_Avg"] <= 2,
        df["Exp_Avg"].between(3, 6),
        df["Exp_Avg"] >= 7
    ],
    [
        "Junior",
        "Mid",
        "Senior"
    ],
    default="Unknown"
)

df["ExperienceLevel"].value_counts(dropna=False)
```

```python
# Salary Bucket

df["SalaryBucket"] = np.select(
    [
        df["Salary_Avg"] <= 300000,
        df["Salary_Avg"].between(300001, 700000),
        df["Salary_Avg"]> 700000
    ],
    [
        "Low",
        "Medium",
        "High"
    ],
    default= "Unknown"
)

df["SalaryBucket"].value_counts(dropna=False)
```

```python
# 
df["SkillCount"] = (
    df["skills"]
    .fillna("")
    .apply(lambda x: len(x.split(","))if x.strip() else 0)
)

df["SkillCount"].describe()
```

```python
# PrimarySkill

def extract_primary_skill(skill_text):
    if not isinstance(skill_text, str):
        return "Unknown"

    s = skill_text.lower()

    if "application programming" in s or "software" in s:
        return "Software Development"
    elif "qa" in s or "testing" in s:
        return "QA / Testing"
    elif "data" in s or "analytics" in s:
        return "Data"
    elif "accounts" in s or "finance" in s:
        return "Finance"
    elif "marketing" in s:
        return "Marketing"
    elif "hr" in s or "human resources" in s:
        return "HR"
    else:
        return "Other"

df["PrimarySkill"] = df["skills"].apply(extract_primary_skill)

df["PrimarySkill"].value_counts()
```

```python
# HighValueRole Flag
df["HighValueRole"] = np.where(
    (df["ExperienceLevel"] == "Senior") &
    (df["SalaryBucket"] == "High"),
    1,
    0
)

df["HighValueRole"].value_counts()
```

```python
df.to_csv(
    "../Data/Processed/featured_job_listings.csv",
    index = False
)
```
