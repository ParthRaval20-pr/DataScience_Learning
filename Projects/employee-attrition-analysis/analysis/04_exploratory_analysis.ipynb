```python
import pandas as pd
```

```python
df = pd.read_csv(
    "../data/processed/featured_employee_data.csv",
    parse_dates=["JoinDate"]
)
```

```python
attrition_rata = df["AttritionFlag"].mean()
```

```python
attrition_by_dept = (
    df.groupby("Department")["AttritionFlag"].mean().sort_values(ascending = False)
)
```

```python
attrition_by_exp = (
    df.groupby("ExperienceLevel")["AttritionFlag"]
      .mean()
      .sort_values(ascending=False)
)
```

```python
salary_stats = (
    df.groupby("Attrition")["Salary"]
      .agg(["mean", "median", "std", "count"])
)
```

```python
performance_attrition = (
    df.groupby("PerformanceBand")["AttritionFlag"]
      .mean()
      .sort_values(ascending=False)
)
```

```python
early_exit_attrition = (
    df.groupby("EarlyTenureExitRisk")["AttritionFlag"]
      .mean()
)
```
