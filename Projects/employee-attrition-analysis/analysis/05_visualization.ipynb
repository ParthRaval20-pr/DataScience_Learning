```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv(
    "../data/processed/featured_employee_data.csv",
    parse_dates=["JoinDate"]
)

attrition_by_dept = (
    df.groupby("Department")["AttritionFlag"].mean().sort_values(ascending = False)
)
```

```python
plt.figure(figsize=(8, 5))
plt.bar(
    attrition_by_dept.index,
    attrition_by_dept.values
)
plt.title("Attrition Rate by Department")
plt.xlabel("Department")
plt.ylabel("Attrition Rate")
plt.xticks(rotation=45)
plt.tight_layout()

plt.savefig("../outputs/figures/attrition_by_department.png")
```

```python
plt.figure(figsize=(8, 5))

sns.barplot(
    x=attrition_by_dept.index,
    y=attrition_by_dept.values
)

plt.title("Attrition Rate by Department")
plt.xlabel("Department")
plt.ylabel("Attrition Rate")

plt.xticks(rotation=45)
plt.tight_layout()

plt.savefig("../outputs/figures/attrition_by_department_seaborn.png")
```

```python
plt.figure(figsize=(7, 5))

sns.boxplot(
    x="Attrition",
    y="Salary",
    data=df
)

plt.title("Salary Distribution: Leavers vs Stayers")
plt.xlabel("Attrition")
plt.ylabel("Salary")

plt.tight_layout()
plt.savefig("../outputs/figures/salary_by_attrition.png")
```

```python
plt.figure(figsize=(8, 5))

sns.scatterplot(
    x="TotalExperience",
    y="Salary",
    hue="Attrition",
    data=df,
    alpha=0.7
)

plt.title("Salary vs Experience (Attrition Highlighted)")
plt.xlabel("Total Experience (Years)")
plt.ylabel("Salary")

plt.tight_layout()
plt.savefig("../outputs/figures/salary_vs_experience.png")
```

```python
plt.figure(figsize=(6, 4))

sns.barplot(
    x="ExperienceLevel",
    y="AttritionFlag",
    data=df,
    order=["Junior", "Mid", "Senior"]
)

plt.title("Attrition Rate by Experience Level")
plt.xlabel("Experience Level")
plt.ylabel("Attrition Rate")

plt.tight_layout()
plt.savefig("../outputs/figures/attrition_by_experience_level.png")
```

```python
plt.figure(figsize=(7, 5))

corr = df[
    ["Age", "YearsAtCompany", "TotalExperience",
     "Salary", "PerformanceRating", "AttritionFlag"]
].corr()

sns.heatmap(
    corr,
    annot=True,
    cmap="coolwarm",
    fmt=".2f"
)

plt.title("Correlation Matrix")
plt.tight_layout()
plt.savefig("../outputs/figures/correlation_heatmap.png")
```
