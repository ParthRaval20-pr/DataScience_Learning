```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv(
    "../data/processed/cleaned_employeedata.csv",
    parse_dates=["JoinDate"]
)

df["SalaryPerExperience"] = df["Salary"] / df["TotalExperience"]
```

```python
df["ExperienceLevel"] = np.select(
    [
        df["TotalExperience"] <= 5,
        df["TotalExperience"].between(6, 12),
        df["TotalExperience"] >= 13
    ],
    ["Junior", "Mid", "Senior"],
    default="Unknown"
)
```

```python
df["AttritionFlag"] = np.where(df["Attrition"] == "Yes", 1, 0)
```

```python
df["PerformanceBand"] = np.where(
    df["PerformanceRating"] <= 2, "Low",
    np.where(df["PerformanceRating"] == 3, "Average", "High")
)
```

```python
df["EarlyTenureExitRisk"] = (df["YearsAtCompany"] <= 2).astype(int)
df["EarlyTenureExitRisk"].value_counts()
```

```python
df.to_csv(
    "../data/processed/featured_employee_data.csv",
    index=False
)
```
