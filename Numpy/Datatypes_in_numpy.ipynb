```python
# numpy Datatypes
import numpy as np
arr = np.array([1,2,3,4,5])
print(arr.dtype)
```

```python
arr = np.array([1.5,2.7,3.9])
print(arr.dtype)
```

```python
# changing datatypes
arr_int = arr.astype(np.int32)
print(arr_int)
print(arr_int.dtype)
```

```python
# Downcasting to save mamory
arr_large = np.array([10000,20000,30000],dtype = np.int64)
arr_small = arr_large.astype(np.int32)
print(arr_small)
print(arr_small.dtype)
```

```python
# mamory usage
arr_int64 = np.array([1,2,3], dtype=np.int64)
arr_int32 = np.array([1,2,3], dtype=np.int32)

print(arr_int64.nbytes)
print(arr_int32.nbytes)
```

```python
# String Data Type
arr = np.array(['apple','bananna','cherry'],dtype='U10')
print(arr)
```

```python
# Complex Numbers
arr = np.array([1 + 2j, 3 + 4j, 5 + 6j], dtype='complex128')
print(arr)
```

```python
# Object Data Type
arr = np.array([{'a': 1}, [1, 2, 3], 'hello'], dtype=object)
print(arr)
```
